```python
ip=data.drop(['broken'],axis=1)
```

```python
ip
```

```python
op=data['broken']
```

```python
#0ne hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer([('team',OneHotEncoder(),[4]),
                     ('provider',OneHotEncoder(),[5])],
                     remainder='passthrough')
```

```python
ip=np.array(ct.fit_transform(ip),dtype=str)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
x_train.shape
```

```python
# standard scalar transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
```

```python
yp
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1=ConfusionMatrixDisplay(cm)
cm1.plot()
plt.show()
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
precision=precision_score(y_test,yp)
f1=f1_score(y_test,yp)
```
